## Export results

```python

import io
import os.path
import PIL.Image as Image
from pathvalidate import sanitize_filename

out = (f"./{sanitize_filename(api_helper.selection['run_title'])}/")
figures = api_helper.selection['run']
id = api_helper.selection['run_root_id']

if os.path.isdir(out):  
    print(out + ' exists.')
else:
    os.mkdir(out)   
    print(out + ' created.')

for image_id in figures:
    image_bytes = api_helper.asdc_api.asdc_download(id, image_id).content
    image = Image.open(io.BytesIO(image_bytes))
    writeout = os.path.join(out, figures[image_id]['display_name'])
    print("-----------------------------------------------------------------------")
    print(f"{figures[image_id]['display_name']}")
    print("-----------------------------------------------------------------------")
    print(f"Write: {writeout}")
    image.save(writeout)
```
